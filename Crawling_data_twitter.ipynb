### Crawler that collects a user's social network information given the user's ID (on this case I'm using username)

```
data_following = tw.Cursor(api.friends, "monotune").items(30) #getting the first 30 following/friends of monotune account
data_followers = tw.Cursor(api.followers, "monotune").items(30) #getting the first 30 followers of monotune account
following = [[following.id, following.screen_name] for following in data_following]
followers = [[followers.id, followers.screen_name] for followers in data_followers]
data = [[following[i][0], following[i][1], followers[i][0], followers[i][1]] for i in range(30)]
user_network = pd.DataFrame(data=data, 
                    columns=['id_following', 'username_following', "id_followers", "username_followers"])
user_network
```

### Crawler that collects the tweets using a set of specified keywords and a geolocation based criteria

```
search_words = "vaksinasi -filter:retweets" #using vaksinasi indonesia keyword and filtered without retweets
date_since = "2021-03-01" #limit tweet greater than March, 01 2021
geo = '-6.97397368996023,107.63019919825788,25km' #using Tel-U as the central of location with a radius of 25km

tweets = tw.Cursor(api.search,
                   q=search_words,
                   since=date_since,
                   geocode=geo).items(30) #getting 30 tweets that contains vaksinasi keywords and specified by date and geolocation

datas = [[tweet.user.screen_name, tweet.user.location, tweet.text] for tweet in tweets]
df_tweet = pd.DataFrame(data=datas, 
                    columns=['username', "location", "tweet"])
df_tweet
```
